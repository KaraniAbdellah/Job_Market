## Get The Data From API

```python
# Import Packages
import requests
import json
from openpyxl import Workbook
import pandas as pd
from pprint import pprint
```

```python
# API-ENDPOINT
numberOfDocument = 10
URL = f"https://api.fda.gov/drug/event.json?limit={numberOfDocument}"

def getData():
    # Load Data
    req = requests.get(url=URL)
    json_data = req.json()
    
    print(json_data["results"][1])

    for i in range(0, numberOfDocument):
        data = json_data["results"][i]

        # Explode drug Column
        df = pd.json_normalize(data=data)
        df = df.explode(["patient.drug"])

        # Normalize the drug column
        df_drug = pd.json_normalize(data=df["patient.drug"])
        df_drug.index = df.index

        # Drop drug column
        df = df.drop("patient.drug", axis=1)
        df = pd.concat([df, df_drug], axis=1)

        # Explode reaction Column
        df_reaction = df.explode("patient.reaction")

        # Normalize reaction Column
        df_reaction_norm = pd.json_normalize(data=df_reaction["patient.reaction"])
        df_reaction_norm.index = df_reaction.index

        # Drop origin reaction column
        df_reaction = df_reaction.drop("patient.reaction", axis=1)
        df_final = pd.concat([df_reaction, df_reaction_norm], axis=1)
        df_final

        # Export it to CSV
        df_final.to_csv("data.csv", index=False) # remove index (0, 1, 2, ... n)


getData()
```
